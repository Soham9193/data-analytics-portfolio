###  Does a higher movie budget lead to higher global box-office revenue?


```python
sns.scatterplot(data=df,x='BudgetUSD',y='Global_BoxOfficeUSD',alpha=0.3)
plt.title("Movie Budget vs Global Box Office collection:")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

### Insight
The scatter plot shows a strong positive relationship between movie budget and global box-office revenue.
Higher-budget movies generally earn more revenue; however, the presence of several high-budget
movies with relatively low earnings indicates that a large budget does not guarantee success.
Other factors such as genre, audience reception, and marketing likely influence performance.


### Is there a relationship between IMDb ratings and Rotten Tomatoes scores?


```python
sns.scatterplot(data=df,x='IMDbRating',y='RottenTomatoesScore',alpha=0.3)
plt.title("IMD Rating vs Rotten tomatoes score:")

plt.show()
```

### Insight
The plot shows a clear positive correlation between IMDb ratings and Rotten Tomatoes scores.
Movies that score highly on IMDb tend to also receive higher Rotten Tomatoes scores,
indicating general agreement between audience-based and critic-based evaluations.


```python
df.columns
```

```python
genre_year_stats=df.groupby (['Genre','ReleaseYear']).agg({
    "Global_BoxOfficeUSD": "mean",
    "BudgetUSD":"mean",
    "IMDbRating":"mean",
    "RottenTomatoesScore":"mean"
}).reset_index()
```

```python
genre_year_stats
```

###  How does average global box-office revenue vary by genre over time?


```python
plt.figure(figsize=(15,8))
sns.lineplot(data=genre_year_stats,x='ReleaseYear',y='Global_BoxOfficeUSD',hue='Genre')
plt.title("Average global box office by genre over years")
plt.show()
```

### Insight
Action, Sci-Fi, and Adventure genres consistently show higher average global box-office collections
across years compared to genres such as Drama and Documentary.
While revenue fluctuates year to year, commercially driven genres demonstrate stronger
long-term earning potential.


```python
plt.figure(figsize=(15,10))
sns.lineplot(data=genre_year_stats.query("Genre == 'Action'"),x='ReleaseYear',y='Global_BoxOfficeUSD',hue='Genre')
plt.title("Average global box office by genre over years")
plt.show()
```

```python
top_actors=df.groupby('LeadActor')["IMDbRating"].mean().sort_values(ascending=False).head(50)
```

### Which lead actors are associated with higher average IMDb ratings?


```python
plt.figure(figsize=(10,4))
top_actors.plot(kind='bar')
plt.title("top 50 actors by AVG IMDB rating")
plt.show()
```

### Insight
The analysis highlights several lead actors whose movies consistently achieve higher
average IMDb ratings.
This suggests that actor selection may influence perceived movie quality,
although ratings are also affected by genre, direction, and storyline.

###  Do higher-rated movies generate more global box-office revenue?


```python
plt.scatter(df['IMDbRating'], df['Global_BoxOfficeUSD'], alpha=0.4)
plt.xlabel('IMDb Rating')
plt.ylabel('Global Box Office (USD)')
plt.title('IMDb Rating vs Global Box Office Revenue')
plt.show()
```

### Insight
The relationship between IMDb ratings and global revenue appears weak.
Several movies with moderate ratings achieve high revenue, suggesting that
commercial success is not solely dependent on audience ratings.


```python
df.columns
```

###  Which numerical features are most strongly correlated with revenue?


```python
plt.figure(figsize=(10,6))
sns.heatmap(
    df[['BudgetUSD','Global_BoxOfficeUSD','IMDbRating',
        'RottenTomatoesScore']].corr(),
    annot=True, cmap='coolwarm'
)
plt.title('Correlation Matrix of Movie Features')
plt.show()
```

### Insight
Budget shows the strongest positive correlation with global box-office revenue.
Ratings exhibit weaker correlations, indicating that financial investment
plays a more significant role in commercial success than critical reception alone.
