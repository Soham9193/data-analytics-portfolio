```python
# examinr the distribution of unoque categotries in categorical columns
catagorical_columns=df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','income_band']].columns

for i in catagorical_columns:
    print(f'value counts for {i}:')
    display(df[i].value_counts())
```

### Univarite analysis

```python
for i , predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','income_band']].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

### Bivariate analysis

```python
for i , predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','income_band']].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

#### histplot of value counts for diffrent occupations

```python
for i in catagorical_columns:
    if i == 'Occupation':
        continue
    plt.figure(figsize=(15,8))
    sns.histplot(df[i])
    plt.title('histogram of occupation counts')
    plt.xlabel(i)
    plt.ylabel("count")
    plt.show()
```

### Numerical Columns

```python
df.columns
```

```python
numeric_columns=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts',
       'Foreign Currency Account', 'Business Lending']

for i,col in enumerate(numeric_columns):
    plt.figure(figsize=(15,8))
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()
```

#### heatmap

```python
correlation_matrix=df[numeric_columns].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title('correlation mateix')
plt.show()
```
